mpl_tookitsの中に3Dプロットのできるライブラリがあったので使ってみた。とりあえずポピュラな2変数関数$z = x^2 + y^2$を実装する

```python
# -*- coding: utf-8 -*-
# numerila solution using scipy.integrate
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
#可視化したグラフを触れるようにする
%matplotlib notebook

#z = x^2 + y^2の可視化を行う。
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)

#XとYを平面に拡張
X, Y = np.meshgrid(x, y)
print(X.shape)
z = X**2 + Y**2
 
#plot3D using matplotlib
fig = plt.figure()
ax = Axes3D(fig)
#ワイヤのみの表示
ax.plot_wireframe(X, Y, z)
#表面をつける
#ax.plot_surface(X, Y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
