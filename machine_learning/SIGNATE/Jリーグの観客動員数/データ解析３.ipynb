```python
#coding utf-8
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
stadium = pd.read_csv("stadium.csv")
sample = pd.read_csv("sample_submit.csv", header=None)
cond = pd.read_csv("condition.csv")
cond_add = pd.read_csv("condition_add.csv")
data_2014 = pd.read_csv("2014_add.csv")
train_add = pd.read_csv("train_add.csv")
y = train["y"]
train = train.drop("y", axis=1)
```

```python
df = pd.concat([train, test])
df.info()
```

```python
df = pd.merge(df, cond, left_on="id", right_on="id")
df.head()
```

```python
df = df.drop(["referee", "home_team", "home_01", "home_02", "home_03", "home_04", "home_05", "home_06", "home_07", "home_08",
                 "home_09", "home_10", "home_11", "away_01", "away_02", "away_02", "away_03", "away_04", "away_05", "away_06", "away_07",
                 "away_08", "away_09", "away_10", "away_11"], axis=1)
df.head()
```

```python
df.info()
```

```python
df = df.drop(["tv", "home_score", "away_score", "away_team"], axis=1)
df.head()
```

```python
stadium = stadium.drop("address", axis=1)
```

```python
df = pd.merge(df, stadium, left_on="stadium", right_on="name")
df.head()
```

```python
df["humidity"] = df["humidity"].apply(lambda x: x.split("%")[0])
df["humidity"] = df["humidity"].astype(float)
```

```python
df = df.drop(["name", "year", "home", "away", ], axis=1)
df.info()
```

```python
df["week"] = df["gameday"].apply(lambda x: list(x)[-2])
df = df.drop("gameday", axis=1)
df.info()
```

```python
df["time"] = df["time"].apply(lambda x: x.split(":")[0])
df.head()
```

```python
df["time"] = df["time"].astype(int)
df.head()
```

```python
df = df.drop("match", axis=1)
```

```python
df = df[["week", "stage", "temperature", "humidity", "time", "stadium", "weather"]]
df.info()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
train = df.iloc[:1721, :]
test = df.iloc[1721:, :]
#train = train.drop("id", axis=1)
#test = test.drop("id", axis=1)
```

```python
train.head()
```

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Activation
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(107, input_dim=107))
    model.add(Activation("sigmoid"))
    
    model.add(Dense(214))
    model.add(Activation("relu"))
    model.add(Dropout(0.5))
    
    model.add(Dense(1))
    
    model.compile(loss="mean_squared_error", optimizer="adam")
    
    return model
```

```python
estimator = KerasRegressor(build_fn=create_model, epochs=100, batch_size=10, verbose=1)
estimator.fit(train, y)
```

```python
pred = estimator.predict(test)
```

```python
pred = pred.astype(int)
pred
```

```python
sample[1] = pred
```

```python
sample.to_csv("submit_deep_new1.csv", index=None, header=None)
```
