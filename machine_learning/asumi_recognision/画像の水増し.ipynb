```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
import sys, os
#リストで結果を返す関数
def get_file(dir_path):
    filenames = os.listdir(dir_path)
    return filenames

# ルックアップテーブルの生成
min_table = 50
max_table = 205
diff_table = max_table - min_table

LUT_HC = np.arange(256, dtype = 'uint8' )
LUT_LC = np.arange(256, dtype = 'uint8' )

# ハイコントラストLUT作成
for i in range(0, min_table):
    LUT_HC[i] = 0
for i in range(min_table, max_table):
    LUT_HC[i] = 255 * (i - min_table) / diff_table
for i in range(max_table, 255):
    LUT_HC[i] = 255

# ローコントラストLUT作成
for i in range(256):
    LUT_LC[i] = min_table + i * (diff_table) / 255
```

```python
iguchi_in_jpg = "resize_iguchi/"
iguchi_out_jpg = "mizumashi_iguchi/"
asumi_in_jpg = "resize_asumi/"
asumi_out_jpg = "mizumashi_asumi/"
pic = get_file(iguchi_in_jpg)
for i in range(len(pic)):
    img = cv2.imread(iguchi_in_jpg + pic[i])
    #print(img)
    high_cont_img = cv2.LUT(img, LUT_HC)
    low_cont_img = cv2.LUT(img, LUT_LC)
    #print(resize_img)
    save_path1 = iguchi_out_jpg + 'out1_{}.jpg'.format(i+1)
    save_path2 = iguchi_out_jpg + "out2_{}.jpg".format(i+1)
    print(save_path1)
    print(save_path2)
    cv2.imwrite(save_path1, high_cont_img)
    cv2.imwrite(save_path2, low_cont_img)
```

```python
pic = get_file(asumi_in_jpg)
for i in range(len(pic)):
    img = cv2.imread(asumi_in_jpg + pic[i])
    #print(img)
    high_cont_img = cv2.LUT(img, LUT_HC)
    low_cont_img = cv2.LUT(img, LUT_LC)
    #print(resize_img)
    save_path1 = asumi_out_jpg + 'out1_{}.jpg'.format(i+1)
    save_path2 = asumi_out_jpg + "out2_{}.jpg".format(i+1)
    print(save_path1)
    print(save_path2)
    cv2.imwrite(save_path1, high_cont_img)
    cv2.imwrite(save_path2, low_cont_img)
```

```python
# ガンマ変換ルックアップテーブル

LUT_G1 = np.arange(256, dtype = 'uint8' )
LUT_G2 = np.arange(256, dtype = 'uint8' )
gamma1 = 0.75
gamma2 = 1.5
for i in range(256):
    LUT_G1[i] = 255 * pow(float(i) / 255, 1.0 / gamma1)
    LUT_G2[i] = 255 * pow(float(i) / 255, 1.0 / gamma2)
```

```python
pic = get_file(asumi_in_jpg)
for i in range(len(pic), 2*len(pic)):
    img = cv2.imread(asumi_in_jpg + pic[i-len(pic)])
    #print(img)
    G1_img = cv2.LUT(img, LUT_G1)
    G2_img = cv2.LUT(img, LUT_G2)
    #print(resize_img)
    save_path1 = asumi_out_jpg + 'out1_{}.jpg'.format(i+1)
    save_path2 = asumi_out_jpg + "out2_{}.jpg".format(i+1)
    print(save_path1)
    print(save_path2)
    cv2.imwrite(save_path1, G1_img)
    cv2.imwrite(save_path2, G2_img)
```
