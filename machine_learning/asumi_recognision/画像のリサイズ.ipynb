```python
def image_resize(im):
    height = 100
    width = 100
    resize_img = cv2.resize(im, (width, height))
    
    return resize_img
```

```python
import sys, os
in_jpg = "out_asumi/"
out_jpg = "resize_asumi/"
#リストで結果を返す関数
def get_file(dir_path):
    filenames = os.listdir(dir_path)
    return filenames
 
pic = get_file(in_jpg)
print(pic[0])
```

```python
in_jpg = "out_iguchi/"
out_jpg = "resize_iguchi/"
pic = get_file(in_jpg)
for i in range(len(pic)):
    img = cv2.imread(in_jpg + pic[i])
    #print(img)
    #resize_img =img.resize((100, 100))
    resize_img = cv2.resize(img, (100, 100))
    #print(resize_img)
    save_path = out_jpg + 'out_{}.jpg'.format(i+1)
    print(save_path)
    cv2.imwrite(save_path, resize_img)
```
